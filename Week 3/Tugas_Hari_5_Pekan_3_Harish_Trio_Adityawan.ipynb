````python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
````

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

````python
fig, ax = plt.subplots(figsize=(13, 13))
palette="ch:r=-.2,d=.3_r"

#code here
with sns.axes_style("white"):
    ax = sns.scatterplot(df['carat'],
                        df['price'],
                        palette = palette,
                        hue = df['clarity'],
                        hue_order = ['I1','SI2','SI1','VS1','VS2','VVS2','VVS1','IF'],
                        edgecolor = 'none',
                        size = df['depth'],
                        sizes=(1, 10))
plt.show()
````

````python
#code here
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)
with sns.axes_style("dark"):
    grid = sns.FacetGrid(df, col="cut", hue="cut", height = 4)
grid.map(hexbin, "carat", "price", extent=[0, 3.0, 0, 20000])
plt.show()
````

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price sehingga dihasilkan output sesuai expected.

````python
df2 = df.sample(n=300, random_state=123)

#code here
#with sns.axes_style("white"):
df3 = df2[['cut','carat', 'depth', 'table', 'price']]
sns.pairplot(df3, hue='cut', markers=['+','o','s','D','x'], corner=True)


plt.show()
````

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

````python
#code here
import numpy as np, pandas as pd; np.random.seed(0)
import seaborn as sns; sns.set(style="white", color_codes=False)
import scipy.stats as stats
from warnings import filterwarnings
filterwarnings('ignore')

kde = sns.jointplot("carat", 
                  "price", 
                  data=df2,
                  kind="kde", 
                  space=0, 
                  color="b")

reg = sns.jointplot("carat", 
                    "price", 
                    data=df2, 
                    kind="reg").annotate(stats.pearsonr)

#hex.annotate(stats.pearsonr)
plt.show()
````
