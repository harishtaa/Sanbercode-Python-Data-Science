Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt
plt.style.use('default')
fig, ax = plt.subplots(figsize=(8,4))


#Masukkan Data ke Axis + Styling
ax.plot(x,cos, linestyle='solid', color='R',label = "Cosx", linewidth=1)
ax.plot(x,cos3, linestyle='dotted', color='G', label = '3 Cosx')
ax.plot(x,sin, linestyle='dashdot', color='B', label = 'SIn(x-90)', linewidth=1)
ax.plot(x,sin2, marker='x',linestyle='none', color='Y', label = '2 Sinx',markersize=3)
ax.legend(loc=10, fontsize = 5)

#Memberikan Axis Label dan Title
ax.set_xlabel('X Label Axis', fontsize = 6)
ax.set_ylabel('Y Label Axis', fontsize = 6)
ax.set_title('Grafik Cos Sin', fontsize = 6)
plt.show()
```
